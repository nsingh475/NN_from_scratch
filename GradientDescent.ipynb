```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Generating Training Data
np.random.seed = 3

X = np.random.rand(100, 1)
noise = 0.5*np.random.rand(100, 1)
Y = 2*X + 3 + noise
```

```python
## Visualizing Data 

plt.scatter(X, Y)
plt.xlabel('X')
plt.ylabel('Y')
```

```python
## setting parameters

W = np.random.rand()
learning_rate = 0.1
num_epochs = 1000
```

```python
def cost_function(W, X, Y):
    m = len(X)
    X= X.flatten()
    Y = Y.flatten()
    y_hat = []
    cost = 0
    for i in range(m):
        y_pred = np.dot(X[i], W)
        y_hat.append(y_pred)
        error = (y_pred - Y[i])**2
        cost += error
    return y_hat, cost*1/(2*m)
```

```python
def partial_derivative(X, y_hat, Y):
    m = len(X)
    X= X.flatten()
    Y = Y.flatten()
    diff = np.subtract(y_hat,Y)
    der = np.dot(X, diff)
    return der/m
```

```python
for epoch in range(num_epochs):
    print(f'-------This is epoch {epoch+1}-------')
    print(f'Weight: {W}')
    y_hat, cost = cost_function(W, X, Y)
    print(f'Cost: {cost}')
    derivative = partial_derivative(X, y_hat, Y)
    print(f'Partial Derivative: {derivative}')
    if round(derivative, 6)==0:
        break
    W = W - learning_rate*derivative
    print(f'Updated Weight: {W}')
```

```python
print('\n----------------Final Results----------------')
print(f'Converged after {epoch+1} epoch')
print(f'Final weight is: {W}')
```
