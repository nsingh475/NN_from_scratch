```python
def softmax(z):
    # Stabilizing softmax by z - np.max(z)
    # Calculating the exponential of z
    exps = np.exp(z - np.max(z, keepdims=True))
    # Returning exp/ Sum(exp)
    return exps / np.sum(exps)

def forward_single(n, hidden):
    inp = np.zeros(input_dim)
    inp[n] = 1
    emb = np.tanh((embedding_weights @ inp) + embedding_bias)
    internal_input = np.concatenate((hidden, emb))
    internal_output = np.tanh((internal_weights @ internal_input) + internal_bias)
    output_output = softmax((output_weights @ internal_output) + output_bias)
    return output_output, internal_output

def forward(x):
    hidden = np.random.randn(hidden_state_dim)
    output_chars = []
    hidden_history = [hidden]
    output_distribution = []
    for n in x:
        o, hidden = forward_single(n, hidden)
        output_chars.append(np.argmax(o))
        hidden_history.append(hidden)
        output_distribution.append(o)
    return output_chars, hidden_history, output_distribution

```

```python
def dtanh(x):
    return 1 - np.tanh(x) ** 2

def cross_entropy(pred, true):
    print('CE', pred, true)
    # True labels X log of predicted labels
    losses = np.multiply(true, np.log(pred))
    # Calculating the mean of cross entropy
    # Multiplying by -ve to get a final +ve value (?)
    total_loss = -np.sum(losses)/true.shape[0]
    return losses, total_loss


internal_weights_grads = []
internal_bias_grads = []
embedding_weights_grads = []
embedding_bias_grads = []
output_weights_grads = []
output_bias_grads = []
losses = []

# Vec shape: 1*n
# Mat shape: n*m
# VM shape: n*m
# so the output would be a matrix rather than an array
def mult(vec, mat):
    vm = np.array([vec for i in range(mat.shape[1])])
    return np.multiply(vm.T, mat)

# Run backprogagation
def backprop(X, Y):
    
    global embedding_weights_grads, embedding_bias
    global internal_weights_grads, internal_bias_grads
    global output_weights_grads, output_bias_grads
    global losses
    
    pred_chars, hidden_vectors, y_hat_list = forward(X) # Run forward propagation
    prev_hidden_grad = None # Initialize previous hidden gradation to None
    
    
    
    # Y: expected output
    # hidden_vectors: hidden layer at t-1
    # hidden_vectors[1:]: hidden layer at t
    # y_hat_list: matrix of probabilities for predicted characters
    # X: input
    ### Iterate through above list after reversing the ordering of each list
    for y, h_in, h_out, y_hat, x_char in reversed(list(zip(Y, hidden_vectors, hidden_vectors[1:], y_hat_list, X))):
        updates = {}
        
        print('--------------')
        ### One hot encoding for character we are focusing on in X
        x = np.zeros(input_dim)
        x[x_char] = 1
        
        ### Redo a bit of the forward pass to pull out intermediate values
        embed1 = (embedding_weights @ x) + embedding_bias # Emded1 is weights * input + bias
        embed2 = np.tanh(embed1) # Activation function
        d_embed = dtanh(embed1) # Derivative of activation for later use
        
        cc = np.concatenate((h_in, embed2)) # Concat hidden layer at t-1
        int1 = (internal_weights @ cc) + internal_bias # Int1 is weights * input + bias
        #int2 = np.tanh(int1) # Activation function
        d_int = dtanh(int1) # Derivative of activation for later use
        
        blah = np.zeros(output_dim)
        blah[y] = 1
        
        #print('!!', y_hat, blah)
        ### Actual backward pass
        out_prob = y_hat[y] # Output probability of correct class
        #losses.append(-np.log(out_prob)) # Append loss to losses
        ll = cross_entropy(np.array(y_hat), np.array([y]))
        losses.append(ll[1])
        #d_softmax_and_loss = out_prob * (1 - out_prob) # Calculate derivative of softmax
        d_softmax_and_loss = y_hat - blah 
        rev_val = np.zeros(output_dim) # Create variable that updates as it is passed back
        #rev_val[y] = d_softmax_and_loss # Set correct labels value in rev_val to the derivative of softmax
        rev_val = d_softmax_and_loss
        output_bias_grads.append(rev_val) # Append derivative of softmax to output bias gradations
        
        #output_weights_grads.append(mult(rev_val, output_weights)) # Append updates to output weights to output weight gradation
        output_weights_grads.append(np.dot(rev_val, output_weights))
        rev_val = rev_val @ output_weights # Update rev_val by multiplying by output weights
        
        # If this is not the last time step then add the sum of the gradients
        ### of the timesteps we have already seen
        if prev_hidden_grad is not None:
            rev_val += prev_hidden_grad
        
        rev_val *= d_int # Update rev_val by multiplying the derivative of activation functions for the internal layer
        
        internal_bias_grads.append(rev_val) # Append rev_val to internal bias gradations
        
        updates['ibg'] = rev_val
        
        internal_weights_grads.append(mult(rev_val, internal_weights)) # Append updates to internal weights to output weight gradation
        
        updates['ibw'] = internal_weights_grads[-1]
        
        rev_val = rev_val @ internal_weights # Update rev_val by multiplying by internal weights
        
        prev_hidden_grad = rev_val[:hidden_state_dim] # Split off first half
        rev_val = rev_val[hidden_state_dim:] # Split off second half
        
        rev_val *= d_embed # Update rev_val by derivative of embeddings
        
        embedding_bias_grads.append(rev_val) # Append rev_val to embeddings grad
        
        updates['ebg'] = embedding_bias_grads[-1]
        
        embedding_weights_grads.append(mult(rev_val, embedding_weights)) # Append updates to embedding weights to output weight gradation 
        updates['ebw'] = embedding_weights_grads[-1]
        
        #print(x_char, '->', y, '||argmax|', np.argmax(y_hat),  '||y_hat|', y_hat)
        #print(updates)
```
